```python
# ========================
# Resize and Crop Images in Train and Test Sets
# ========================

# Importing Image from PIL for image manipulation
from PIL import Image
import os

# Paths for training and test datasets
paths = ['asl_data_dmn/asl_alphabet_train/asl_alphabet_train/', 'asl_data_dmn/asl_alphabet_test/asl_alphabet_test/']

# Set the desired size to which images should be rescaled
size = 50  # Rescale size

# List of directories (alphabets) and adding 'nothing' category
list_dirs = list(range(65, 65 + 26))  # A-Z in ASCII values
list_dirs.insert(0, 'nothing')  # Add 'nothing' category to the list

# Variable to track the progress of scaling
completed_percentage = 0.0

# Loop through the directories (train and test sets)
for path in paths:
    for i in list_dirs:
        # Skip the 'nothing' category
        if i == 'nothing':
            continue

        # Get the list of files in the current class (alphabet) directory
        dirs = os.listdir(path + chr(i) + '/')

        # Function to resize and crop images
        def resize():
            global completed_percentage
            # Loop through all files in the directory
            for item in dirs:
                # Build the full file path
                file_path = os.path.join(path, chr(i), item)

                # Check if it's a file
                if os.path.isfile(file_path):
                    # Open the image file
                    im = Image.open(file_path)
                    f, e = os.path.splitext(file_path)

                    # Crop the image to remove any blue borders (adjust the coordinates as needed)
                    im = im.crop((2, 2, 197, 197))

                    # Resize the image to the specified size using LANCZOS filter for better quality
                    imResize = im.resize((size, size), Image.LANCZOS)

                    # Save the resized image in JPEG format with maximum quality
                    imResize.save(f + '.jpg', 'JPEG', quality=100)

                    # Update and display the completion percentage
                    completed_percentage += 1 / (75627 + 5400)
                    print('\rScaling:', round(completed_percentage * 100, 3), '%', end='')

        # Call the resize function for each directory
        resize()

# Print success message after completion
print('\nSuccess')
```

Input: Specifies the input shape for the model.
Conv2D: A 2D convolutional layer for feature extraction in images.
Dense: Fully connected (dense) layer, typically used for the final classification output.
Flatten: Converts the 2D output from convolutional layers to 1D for fully connected layers.
Dropout: A regularization technique to randomly drop neurons during training, helping prevent overfitting.
MaxPooling2D: Reduces the spatial dimensions (height and width) of the feature maps.
atchNormalization: Normalizes the activations of the previous layer to improve the model's convergence.

This set of layers can help to create a robust CNN architecture for image classification.


```python
# Some ANN function imports
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Dropout, MaxPooling2D, BatchNormalization
from tensorflow.keras.models import Model
```

Input Layer:

Takes in images of size 50x50 with 3 color channels (RGB).
First Convolutional Block:

Conv2D with 32 filters and a kernel size of (3, 3).
BatchNormalization for stabilizing training.
MaxPooling2D with a pool size of (5, 5) and stride of (5, 5), which reduces spatial dimensions.
Dropout of 20% to reduce overfitting.
Second Convolutional Block:

Conv2D with 64 filters and the same kernel size of (3, 3).
BatchNormalization and MaxPooling2D as in the first block.
Dropout increased to 30% for further regularization.
Third Convolutional Block:

Conv2D with 128 filters.
BatchNormalization and MaxPooling2D as in the previous blocks.
Dropout increased to 40% for even more regularization.
Fully Connected Layers:

Flatten layer to flatten the output of the final convolutional block into a 1D vector.
Dense layer with 128 neurons and ReLU activation.
Dropout of 30% after the dense layer to prevent overfitting.
Dense output layer with a softmax activation function for multi-class classification, with the number of units equal to the number of classes in the dataset (train_generator.num_classes).
This architecture seems well-balanced for a classification task involving images, with increased filters, dropout, and batch normalization for better performance and generalization.

```python
# Designing our CNN
i = Input(shape=(IMAGE_SIZE[0], IMAGE_SIZE[0], 3))
x = Conv2D(32, (3, 3), activation='relu', padding='same')(i)  # Increased filters & kernel size
x = BatchNormalization()(x)
x = MaxPooling2D(pool_size=(5, 5), strides=(5, 5), padding='same')(x)
x = Dropout(0.2)(x)  # Enabled dropout

x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)  # Increased filters
x = BatchNormalization()(x)
x = MaxPooling2D(pool_size=(5, 5), strides=(5, 5), padding='same')(x)
x = Dropout(0.3)(x)  # Enabled dropout

x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)  # Increased filters
x = BatchNormalization()(x)
x = MaxPooling2D(pool_size=(5, 5), strides=(5, 5), padding='same')(x)
x = Dropout(0.4)(x)  # Enabled dropout

x = Flatten()(x)
x = Dense(128, activation='relu')(x)  # Increased neurons
x = Dropout(0.3)(x)  # Increased dropout to prevent overfitting
x = Dense(K, activation='softmax')(x)

model = Model(i, x)
```
